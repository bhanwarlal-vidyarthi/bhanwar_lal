# Objective

Perform exploratory data analysis (EDA) on YouTube channel dataset

## Data Description

Dataset contains 5000 YouTube channels:
Columns present:Rank, grade, channel name, video uploads, subscribers,video views.


```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('top-5000-youtube-channels.csv')
data
```

### Exponential values to decimal values

```python
pd.options.display.float_format = '{:.2f}'.format # Converting exponential value to 2 decimal values
```

```python
data.describe()
```

### Replacing '--' Values with NaN

```python
data.head(20)
```

```python
data['Subscribers'] = data['Subscribers'].replace('--', np.nan, regex = True) # To replace '--' with NaN
```

```python
data.head(20) 
```

### Removing Null Values

```python
data.isnull().sum()
```

```python
per_missing = data.isnull().sum()*100/len(data) # Percentage of Missing Values
```

```python
per_missing
```

```python
sns.heatmap(data.isnull()) #Visualization of Null values in data
```

```python
data.dropna(axis = 0, inplace = True) # Dropping missing values
```

```python
sns.heatmap(data.isnull()) # Showing no Missing values present
```

### Data Cleaning in Rank Column

```python
data # Rank Column contains string and Commas
```

```python
data['Rank'] = data['Rank'].str[0:-2] #Slicing to remove string from rank column
```

```python
data['Rank'] = data['Rank'].str.replace(',', '').astype('int')
```

```python
data # No, string and commas in Rank column
```

### Converting Each Column to integer

```python
data.dtypes
```

```python
data['Video Uploads'] = pd.to_numeric(data['Video Uploads'],errors='coerce') # Cleaning of vedio upload column
```

```python
data['Video Uploads'] = data['Video Uploads'].astype('Int64')
```

```python
data['Subscribers'] = pd.to_numeric(data['Subscribers'],errors='coerce') #Cleaning of Subscribers column
```

```python
data.dtypes
```

```python
data['Grade'].unique()
```

```python
data['Grade'] = data['Grade'].map({'A++ ':1, 'A+ ':2, 'A ':3, 'A- ':4, 'B+ ':5})
```

```python
data['Grade'] = data['Grade'].astype('Int64')
```

```python
data.dtypes
```

### Find Average Views of Each Channel

```python
data.columns
```

```python
data['Avg_views'] = data['Video views']/data['Video Uploads']
```

```python
data # Average view can be found in Avg_view column
```

### Top 5 Channels uploade highest vedios

```python
data.sort_values(by = 'Video Uploads', ascending=False).head()
```

### Grade which has maximum number of Video Uploads

```python
sns.barplot(x = 'Grade', y = 'Video Uploads', data = data)
```

### Grade Having higher number of views

```python
sns.barplot(x = 'Grade', y = 'Avg_views', data = data)
```

### Grade Having higher number of views

```python
sns.barplot(x = 'Grade', y = 'Subscribers', data = data)
```

# Conclusion
- Subscribers column contained missing values (~7.7%)
- Some channels have extremely high uploads but low average views
- Grade categories show variation across uploads, views, and subscribers
- Average views metric provides better performance insight than total views
